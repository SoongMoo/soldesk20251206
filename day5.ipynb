```python
import numpy as np
# 조건 논리를 배열 연산으로 표현하기
xarr = np.array([1.1, 1.2, 1.3, 1.4, 1.5])
yarr = np.array([2.1, 2.2, 2.3, 2.4, 2.5])
cond = np.array([True, False, True, True, False])
print(xarr[cond]) # True
print(yarr[cond]) # True
result = []
# 배열, 튜플, 리스트 여러개를 한번에 사용하려면 zip()함수를 사용
# 일반 반복문
for x, y, c in zip(xarr, yarr, cond):
    if c:
        result.append(float(x))
    else:
        result.append(float(y))
print(result) #  리스트
# 파이썬 반복문
result = [float(x if c else y) for x, y, c in zip(xarr, yarr, cond)]
print(result) #  리스트
# np.where()사용 
result = np.where(cond, xarr, yarr )
print(result) # 배열
# arr배열의 값이 0보다 크면 2를 , 0보다 작으면 작으면 -2로 대체하시오
rng = np.random.default_rng(seed=12345)
arr = rng.standard_normal((4, 4)) # 평균 0, 표준편차 1인 정규분포(음수, 양수)
print(arr)
result = np.where(arr > 0 , 2, -2)
print(result)
```

```python
# 누락된 데이터 처리
data = pd.Series([1, np.nan, 3.5, np.nan, 7])
print(data)
print(data.isna())
print(data.dropna()) # 누락값 제외하고 출력
print(data[data.notna()]) # 누락값 제외하고 출력
```

```python
data = pd.DataFrame([[1., 6.5, 3.], [1., np.nan, np.nan],
                     [np.nan, np.nan, np.nan], [np.nan, 6.5, 3.]])
print(data)
print(data.dropna()) # 행에 누락 값이 없는 것만
print(data.dropna(how="all")) # axis='index' 생략 # 하나의 행에 모두 nan인걸 제외
print(data.dropna(how="all", axis='index'))  # index가 NaN일 때는 제거
print(data.dropna(how="all", axis='columns')) # 열의 값이 모두 NaN 인 경우에만 제거
```

```python
print(data)
print(data.dropna(how="any")) # index에 적어 하나가NaN라면 제거
print(data.dropna(how="any", axis='index'))
print(data.dropna(how="any", axis='columns'))
```

```python
print(data)
print(data.mean()) # 열의 평균값
print(data.fillna(data.mean())) # 열의 평균값으로 채우기
print(data.fillna(method='ffill')) # 바로 앞의 데이터로 채워라
print(data.fillna(method='bfill')) # 바로 밑에 있는 데이터로
print(data.fillna(method='bfill', limit=2)) 
# 연속된 NaN 값이 있을 때, 최대 2개까지만 앞의 값으로 채워줍니다
```

```python
data = pd.DataFrame({"k1": ["one", "two"] * 3 + ["two"],
                     "k2": [1, 1, 2, 3, 3, 4, 4]})
print(data)
print(data.duplicated()) # Ture 이면 중복 , 부울 index
print(data[data.duplicated()]) # 부울 index # 중복된행 만 출역
```

```python
print(data)
data["v1"] = range(7)
print(data)
print(data.drop_duplicates(subset=["k1"])) # 중복된 행은 제거 해서 한번만 출력
print(data.drop_duplicates(subset=["k1", "k2"]))# 중복된 행은 제거 해서 한번만 출력
print(data.drop_duplicates(subset=["k1", "k2"], keep="last"))
```

```python
data = pd.Series([1., -999., 2., -999., -1000., 3.])
print(data)
print(data.replace(-999, np.nan))
print(data.replace([-999,-1000], np.nan))
print(data.replace([-999,-1000], [np.nan, 0])) # -999는 1000으로  nan은 0으로
print(data.replace({-999 : np.nan, -1000: 0}))  # -999는 1000으로  nan은 0으로
```

```python
# group by
df = pd.read_csv("https://raw.githubusercontent.com/SoongMoo/data-analysis/refs/heads/main/examples/gapminder.tsv"
                 , sep="\t")
df
```

```python
df.groupby('year')['lifeExp'].mean() # 같은 년도끼리 묶어라.
df.groupby('year').lifeExp.min()
df.groupby('year')['gdpPercap'].mean()
df.groupby('year').gdpPercap.max()
```

```python
df.groupby(['continent', 'country']).lifeExp.mean()
df.groupby(['continent', 'country']).lifeExp.min()
df.groupby(['continent', 'country']).lifeExp.max()
df.groupby(['continent', 'country']).gdpPercap.min()
df.groupby(['year', 'continent'])[['lifeExp','gdpPercap']].min()
```

```python
import seaborn as sns
# 타이타닉 데이터셋 불러오기
df = sns.load_dataset('titanic')
df
```
