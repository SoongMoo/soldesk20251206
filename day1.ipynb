```python
# format함수를 이용한 포맷팅 : format()
# "a" : 문자열, a :변수 , a() : 함수 : print()
"이숭무의 나이는 20살이고 키는 175.3입니다"
name = "이숭무"
age = 20
height = 175.3
print("%s의 나이는 %d살이고 키는 %f입니다" % (name, age, height))
print("{1}의 나이는 {0}살이고 키는 {2}입니다".format(age, name, height))
#                                                    0    1      2 : index
print("{0}의 나이는 {1}살이고 키는 {2}입니다".format( name, age, 175.3))
```

```python
# format함수에 있는 이름 사용하기
print("{0}의 나이는 {1}살이고 키는 {2}입니다".format( name, age, 175.3))

print("{0}의 나이는 {1}살이고 키는 {h}입니다".format( name, age, h=175.3))
#                                                     0    1      x:없다

# 오류 : 사용하면 안됨 
#print("{0}의 나이는 {a}살이고 키는 {1}입니다".format( name, a = age, 175.3))
#                                                     0      x       
# positional argument follows keyword argument

# 이름이 중간에 들어갈 수 없다
print("{0}의 나이는 {a}살이고 키는 {h}입니다".format( name, a = age, h = 175.3))
# 이름은 오른쪽 부터 왼쪽으로 순차적으로 부여 해줘야 한다. 
```

```python
# 정렬 : > (오른쪽), <(왼쪽) , ^(가운데)
print("{0:>10}의 나이는 {a:>10}살이고 키는 {h:>10}입니다".format( name, a = age, h = 175.3))
print("{0:<10}의 나이는 {a:<10}살이고 키는 {h:<10}입니다".format( name, a = age, h = 175.3))
print("{0:^10}의 나이는 {a:^10}살이고 키는 {h:^10}입니다".format( name, a = age, h = 175.3))
print("{0:#^10}의 나이는 {a:$^10}살이고 키는 {h:%^10}입니다".format( name, a = age, h = 175.3))
```

```python
#f 문자 포매팅
print("이숭무의 나이는 20살이고 키는 175.3입니다")
name = "이숭무"
age = 20
height = 175.3
print(f"{name}의 나이는 {age}살이고 키는 {height}입니다")
# 정렬
print(f"{name:10}의 나이는 {age:10}살이고 키는 {height:10}입니다")
print(f"{name:>10}의 나이는 {age:>10}살이고 키는 {height:>10}입니다")
print(f"{name:<10}의 나이는 {age:<10}살이고 키는 {height:<10}입니다")
print(f"{name:^10}의 나이는 {age:^10}살이고 키는 {height:^10}입니다")
print(f"{name:$^10}의 나이는 {age:#^10}살이고 키는 {height:@^10}입니다")
```

```python
print(f"{name:10}의 나이는 {age:->10}살이고 키는 {height:10}입니다")
```

```python
str1 = "Life is too short"
#       0123456789...
# 문자열의 크기
print(len(str1))
# 특정문자의 갯수
print(str1.count("o"))
print("Life is too short".count("o"))

print(str1[1:1+3])
# 문자의 index를  찾기
# i가 처음 나오는  index
idx = str1.index('i') # 1
print(idx)
print(str1[idx : idx + 3])

# i가 두번째에 있는 index
idx = str1.index('i', 2 ) 
idx = str1.index('i', 1 + 1 ) 
idx = str1.index('i', str1.index('i') + 1 ) 
print(str1[idx : idx + 6])

# too가 있는 index를 찾으시오.
idx = str1.index('too') # 8
a = 'too'
print(idx) # too의 첫번재 문자의  index를 가져온다 ('t'의 index)
print(str1[8 : 8 + 3] ) # too출력
print(str1[idx : idx + len(a)] )
```

```python
str1 = "Life is too short"
#                 1
#       0123456789012
idx =  str1.index('s', str1.index('s') + 1 ) # index : left-index
print("idx = ",idx) # index 출력
print(str1[idx:]) # slicing을 이용해서 문자 출력 
idx =  str1.rindex('s') # rindex : right-index
print(idx) 
print(str1[idx:]) # slicing을 이용해서 문자 출력 
print(str1[idx]) # indexing을 이용해서 문자 출력
```
