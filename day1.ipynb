```python
# 02장 파이썬 프로그래밍의 기초, 자료형
# 리터널 자료형 : 숫자 자료형  : 정수 자료형 : 10, 5, 130 : 정수 리터널
#                            : 실수 자료형 : 1.5, 10.5  : 실수 리터널
#                문자 자료형 
print( 10 + 5 ) # 정수 + 정수
a = 10 # a는 변수 : 정수 리터널 변수
b = 5  # b는 변수 : 정수형 변수
print( a + b )
# 변수란 오직 하나의 값만을 가진다
a = 7 # 10 대신에 7로 바뀜
print( a + b )
# 숫자리터널의 연산자 : +, -, *, / : 사직연산자
print( a - b )
print( a * b )
print( a / b )

a = 10.5 # 7대신 10.5로 바뀜
         # 실수 리터널 변수
print( a + b ) # 10.5 + 5 # 실수 + 정수
# 나머지 연산자 
a = 10
b = 3
print( a / b ) # 나누기
print( a % b ) # 나머지 연산자
# 몫 연사자
print( a // b ) # 몫 연산자
# 제곱승
print( a * b ) #곱셈
print( a ** b ) # 제곱승
# 숫자 연산자 : +, -, / , *, %, //, **

#변수 = 값
c = 10
c = a # a값을 c에 저장
print(a)
print(c)
a = a + 1 # a = 10 + 1
print( a )
a = a + 1 # a = 11 + 1
print( a)
a += 1 # a = a + 1
b -= 1 # b = b - 1
a += 5 # a = a + 5

# 복합 연산자
# +=, -=, *=, /=, %=, //=, **=
```

```python
# 문자열 인덱싱
# 문자열의 index를 이용해서 원하는 문자(글자하나)를 가져오는 것
# index는 각 문자에 부여된 고유의 번호( 0 부터 1씩 증가)
str1 = 'Life is too short, You need Python'
#                 1         2         3
#       0123456789012345678901234567890123 : 양수 -->
#           3         2         1      
#      -4321098765432109876543210987654321 : 음수 <---                               
print(str1[3])
print(str1[-31])

print(str1[23])
print(str1[-11])

print(str1[33])
print(str1[-1])

print(34-1) # 33
print(len(str1)) # 34
print(str1[34-1]) # len(str1) <==> 34
print(str1[len(str1)-1])

print(str1[0])
print(str1[-34])
print(str1[-len(str1)])
```
