```python
# 클래스변수 초기화
class Calculator:
    def setData(self, first, second):
        self.first = first
        self.second = second
    def add(self):
        self.result = self.first + self.second
        return self.result
    def div(self):
        self.result = self.first / self.second
```

```python
# 객체가 생성될 때 클래스변수를 초기화하자. (클래스변수를 초기화하는 메서드를 실행안함)
# 생성자
class Calculator:
    def __init__(self, first, second):#객체가 생성이 될 때 자동으로 실행되는 메서드
        self.first = first            #생성자
        self.second = second
    def add(self):
        self.result = self.first + self.second
        return self.result
    def div(self):
        self.result = self.first / self.second
```

```python
class FourCal(Calculator):
    def mul(self):
        self.result = self.first * self.second
    def sub(self):
        self.result = self.first - self.second
    def div(self): # override # 재정의 
        if self.second == 0:
            self.result = 0
        else:
            self.result = self.first / self.second
```

```python
c = FourCal(3, 0)
c.div() # 3 / 0
print(c.result)
```

```python
# 스크래핑
import requests
# 크롤링
from bs4 import BeautifulSoup
# 데이터분석
import pandas
# 시각화
import matplotlib as mpl
import matplotlib.pyplot as plt

# 스크래핑
url = "https://www.weather.go.kr/w/observation/land/city-obs.do"
source = requests.get(url)

# 크롤링
soup = BeautifulSoup(source.content, 'html.parser')
table = soup.find('table',{'class':'table-col'})

# 크롤링한 데이터 저장
data =[]
for tr in table.find_all('tr'):
    tds = list(tr.find_all('td'))
    for td in tds:
      if td.find("a"):
        point = td.find("a").string
        temp = tds[5].string
        humidity = tds[10].string
        data.append([point,temp,humidity])

data
# 데이터를 분석
# 데이터 저장
with open("weather.csv" , "w" , encoding="utf-8") as f:
    f.write('지역,온도,습도\n')
    for l in data:
        f.write(f"{l[0]},{l[1]},{l[2]}\n")
#테이터 처리
df = pandas.read_csv("weather.csv",  encoding='utf-8', index_col='지역')
#print(df)
city_df = df.loc[['서울','인천','대전','대구','광주','부산','울산']]
#print(city_df)
#시각화
font_name = mpl.font_manager.FontProperties(fname="C:\Windows\Fonts\malgun.ttf").get_name()
mpl.rc('font', family=font_name)
ax = city_df.plot(kind='bar', title='weather', figsize=(12,4))
plt.show()
```

```python
# Boolean Indexing
names = np.array(["Bob", "Joe", "Will", "Bob", "Will", "Joe", "Joe"])
data = np.array([[4, 7], [0, 2], [-5, 6], [0, 0], [1, 2],[-12, -4], [3, 4]])
print(names)
print(data)
print(names == "Bob") # [ True False False  True False False False]
print(data[[ True, False, False,  True, False, False, False]])
print(data[names == "Bob"])
print(data[names == "Bob", 1]) # indexing
print(data[names == "Bob", 1:]) # slicing
# Bob이 아닌 index에 해당하는 것만 다 가져오시오
# [False,  True,  True, False,  True,  True,  True]
print(data[~(names == "Bob")])
print(data[names != "Bob"])

cond = names == "Bob"
print(data[cond])
print(data[~cond])

mask = (names == "Bob") | (names == "Will")
print(mask)
print(data[mask])
print(names[mask])
print(data[[ True, False,  True,  True,  True, False, False]])

print(data[data < 0])  # [ -5 -12  -4]
data[data < 0] = 0
print(data)
print([names != "Joe"]) # [ True, False,  True,  True,  True, False, False]
print(data[names != "Joe"]) #
```

```python
# boolean index를 이용해서 데이터 분석
import pandas as pd
rainfall = pd.read_csv("https://raw.githubusercontent.com/SoongMoo/data-analysis/refs/heads/main/data/Seattle2014.csv")
#print(rainfall)
#print(rainfall['PRCP'])
#  배열로 받아옴
inches =  np.array(rainfall['PRCP'] / 254 )
inches
# 0.5인치이상 온 날은 몇일입니까?
#True :1 , False :0
print(np.sum(inches >= 0.5))
# 강수량이 0.5에서 0.8 사이인 날은 몇일이나 되나요?
print(np.sum((inches >= 0.5) & (inches <= 0.8)))
#강수량이 0.5보다 작고 0.8보다는 많이 온날은?
print(np.sum(~((inches >= 0.5) & (inches <= 0.8))))
print(np.sum((inches < 0.5) | (inches > 0.8)))
#비가 안온 날
print(np.sum(inches == 0))
#비가 온 날
print(np.sum(inches != 0))  
```
