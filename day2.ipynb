```python
str1 = "Life is too short,\n You need python"
# indexing, slicing
print(str1)
str1 = """Life is too short,
You need python"""
print(str1)
print(str1[0 : 0 + 4])
print(str1[12 : 12 + 5]) # slicing 
# "a" :문자열, a : 변수 , a() 함수 : print(), format(), find(), index() 
```

```python
# 문자열 함수
#  모두 대문자 변경
str1 = "Life is too short"
print(str1.upper())
#  모두 소문자
print(str1.lower())
```

```python
str1 = "    highland0       "
#공백문자를 지우기
print(str1.strip() + "이숭무")
print(str1.rstrip() + "이숭무")
print(str1.lstrip() + "이숭무")
str1 = "    high    land0    "
print(str1.strip() + "이숭무") # 중간에 있는 공백문자는 지워지지 않는다.

str1 = "abhighland0ab"
print(str1.strip("ab"))

str1 = "ab    highland0  ab"
print(str1.strip("ab").strip())
```

```python
# 문자열 변경하기
str1 = 'pithon'
print(str1[0]+'y'+str1[2:])
print(str1.replace("i","y"))

str1 = 'high land0'
print(str1.replace(" ",""))

str1 = "Life is too short"
result = str1.replace("Life" , "Your leg")
print(result) # 변경된 데이터를 result에 저장
print(str1) # 원본 데이터는 변경 되지 않는다.
```

```python
# 쪼개기
str1 = "Life is too short"
print(str1)
print(str1.split()) # 공백문자 기준으로 분리한다.
# ['Life', 'is', 'too', 'short'] : 리스트
str1 = "Life:is:too:short" #  콜론(:)을 기준으로 분리한다.
print(str1.split(":"))
```

```python
str1 = "Life:is-too/short"
import re
print(re.split('[:/-]',str1))
# ['Life', 'is', 'too', 'short']  : 리스트
```

```python
# 자료형
# 리터널 자료형 : 숫자 리터널 : 정수 리터널 : 10
#                             실수 리터널 : 1.5
#                문자 리터널 : "", '', """""", ''''''
# 리터널 자료형을 변수에 저장 할 때는 변수에 하나만 저장이 된다.
a = 10
b = 10.5
str1 = "Life"
str2 = 'is'
str3 = 'too'
# 하나의 변수에 서로 다른 특징을 가진 여러개의 데이터를 저장하고 싶다.
# 리스트 자료형 : []
str1[1] # indexing
l = ['Life', 'is', 'too'] # 리스트
print(l)
print(a)
print(b)
print(str1)
```

```python
#  리스트 표기법
l2 = [] # 빈 리스트
l3 = [1, 2, 3, 4, 5] # 정수형 리스트 
                     # 여러개의 데이터를 구분하기 위해서 곰마(,)를 사용
l4 = [10.5, 2.5, 12.3] # 실수형 리스트
l5 = ['이숭무', 'apple' ] #문자열 리스트
l6 = [1,2,3, '이숭무', 2.4, 6.7] # 혼홥형 리스트
l7 = [ [1,2,3] , [4,5,6] ] # 리스트를 포함한 리스트
l8 = [ 7,8,9, [1,2,3] ,'이숭무', [4,5,6] ] # 리스트를 포함한 혼합형 리스트
```

```python
# 리스트 연산자 : + : 연결연산자, * : 반복연산자
l1 = [1,2,3]
l2 = [4,5,6]

result = l1 + l2
print(result)
result = l1 * 3
print(result)
```
