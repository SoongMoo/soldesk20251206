```python
#요소 삭제 
# del명령 이용 : index를 사용
l = ['Life', 'is', 'too', 'short']
print(l) # ['Life', 'is', 'too', 'short']
#              0      1     2       3
del l[2] # indexing
print(l) # ['Life', 'is', 'short']
del l[1:] # slicing
print(l)

# 요소를 이용해서 삭제
l = ['Life', 'is', 'too', 'short']
l.remove('is')
print(l) #['Life', 'too', 'short']
# l.remove('Life', 'too')  # one argument (2 given)
# l.remove(['Life', 'too']) #요소인 리스트를 지워라.list.remove(x): x not in list
#remove()도 요소를 하나씩 지운다.

# 한번에 모두 삭제
l.clear()
print(l)

# 데이터를 추출 후 삭제 : 비복원 추출
l = ['Life', 'is', 'short', 2, 3]
# 마지막에 있는 데이터를 추출 후 삭제
result = l.pop()
print(result)
print(l) # ['Life', 'is', 'short', 2]
#              0     1       2     3
# 원하는 위치(index)의 데이터를 추출 후 삭제
result = l.pop(1)
print(result)
print(l)
```
